```python
footagepivot=pd.pivot(data=df_footage,index='Site Name',columns='Month',values='Footage Penalty').reset_index()
footagepivot=footagepivot.fillna(0)
footagepivot=footagepivot[['Site Name','Apr','May','June','July']]
footagepivot
```

```python
footagepivot['Apr']=footagepivot['Apr'].replace(['-','--','----','-----'],0)
footagepivot['May']=footagepivot['May'].replace(['-','--'],0)
footagepivot['June']=footagepivot['June'].replace(['NA ','-','--','---'],0)
footagepivot['July']=footagepivot['July'].replace(['-','--','----'],0)
```

```python
import numpy as np
```

```python
footagepivot.info()
```

```python
footagepivot['Total']=(footagepivot['Apr']+footagepivot['May']+footagepivot['June']+footagepivot['July'])
```

```python
todrop2=footagepivot[footagepivot['Total']==0]

footagepivot=footagepivot.drop(todrop2.index)
```

```python
footagepivot['Site Name'].nunique()
```

```python
footagepivot
```

```python
footagepivot['Mean']=(footagepivot['Apr']+footagepivot['May']+footagepivot['June']+footagepivot['July'])/4
footagepivot['Std']=np.std(footagepivot[['Apr','May','June','July']],axis=1).round(2)
footagepivot['Bucket']=pd.cut(footagepivot['Mean'],bins=[0,99,199,499,999,1999,2999,4999,9999,19999,24999],labels=['<100','100-200','200-500','500-1000','1000-2000','2000-3000','3000-5000','5000-10000','10000-20000','>20000'])
footagepivot['CoefOfVar']=(footagepivot['Std']/footagepivot['Mean']).round(2)
```

```python
footagepivot
```

```python
footagepivot['Regularity']=pd.cut(footagepivot['CoefOfVar'],bins=[-1,0.05,0.5,1,1.5,2],labels=["Very Regular","Regular","Tolerant","Irregular","Very Irregular"])
```

```python
footagepivot
```

```python
footagepivot['Regularity'].value_counts()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression() 
```

```python
X=np.array([[1,2,3,4]]).reshape(-1,1)
X
```

```python
for i in range(0,len(footagepivot[['Apr','May','June','July']]+1)):
    X=np.array([[1,2,3,4]]).reshape(-1,1)
    Y=np.array(footagepivot[['Apr','May','June','July']].iloc[i]).reshape(-1,1)
    model.fit(X,Y)
    footagepivot["Slope"].iloc[i]=(model.coef_[0][0]).round(2)
```

```python
footagepivot
```

```python
conditions = [
    (footagepivot['CoefOfVar'] >=-0.05) & (footagepivot['CoefOfVar'] <=0.05),
    ((footagepivot['CoefOfVar'] <-0.05) & (footagepivot['CoefOfVar']>=-0.5)) | ((footagepivot['CoefOfVar'] >0.05) & (footagepivot['CoefOfVar'] <0.5)) & footagepivot['Slope'] > 0,
    ((footagepivot['CoefOfVar'] <-0.05) & (footagepivot['CoefOfVar']>=-0.5)) | ((footagepivot['CoefOfVar'] >0.05) & (footagepivot['CoefOfVar'] <0.5)) & footagepivot['Slope'] < 0,
    ((footagepivot['CoefOfVar'] <-0.5) | (footagepivot['CoefOfVar'] >=0.5)) & footagepivot['Slope'] > 0,
    ((footagepivot['CoefOfVar'] <-0.5) | (footagepivot['CoefOfVar'] >=0.5)) & footagepivot['Slope'] < 0]
values = [ 'Stable','Increasing','Decreasing','Agg Increasing','Agg Decreasing']
footagepivot['Data'] = np.select(conditions, values)
```
