```python
import pandas as pd
```

```python
lvb=pd.ExcelFile('LVB Overall penalty data Q2 & Q3 2021.xlsx')
```

```python
df1=pd.read_excel(lvb,'MCU Q2-21')
df1=df1.rename(columns=df1.iloc[0]).drop(df1.index[0])
df_mcu=df1[['SiteName','Month','Penalty amount']]
df_mcu=df_mcu.iloc[:,0:4]
df_mcu['NC-30 Penalty']=df_mcu['Penalty amount']
df_mcu['Month']=pd.DatetimeIndex(df_mcu['Month']).month
df_mcu.drop('Penalty amount',axis=1,inplace=True)
df_mcu
```

```python
df_mcu['ATM ID']=df_mcu['SiteName']
df_mcu.drop('SiteName',axis=1,inplace=True)
```

```python
pd.set_option('display.max_columns',None)
df_equi=pd.read_excel(lvb,'Equip. failed case Q2-21')
df_equi['Equipment failure Penalty']=df_equi['Penalty amount']
df_equi
df_equi=df_equi[['ATM ID','Equipment failure Penalty','Open Date and time']]
df_equi=df_equi.iloc[:,0:4]
df_equi['Month']=df_equi['Open Date and time']
df_equi['Month']=pd.DatetimeIndex(df_equi['Month']).month
df_equi.drop('Open Date and time',axis=1,inplace=True)
df_equi.head()
```

```python
df3 = pd.read_excel(lvb, 'Footage tracker Q2-Q3')
df_footage=df3.rename(columns=df3.iloc[0]).drop(df3.index[0])
footage=df_footage[['ATM ID','MONTH','Penalty amount']]
footage['Footage Penalty']=footage["Penalty amount"]
footage['Month']=footage['MONTH']
footage['Month']=pd.DatetimeIndex(footage['Month']).month
footage.drop(['MONTH','Penalty amount'],axis=1,inplace=True)
```

```python
footage.head()
```

```python
df4 = pd.read_excel(lvb, 'PM penalty Q2-21')
df_pmq2=df4.rename(columns=df4.iloc[0]).drop(df4.index[0])
```

```python
pmq2=df_pmq2[['ATM ID','Penalty amount']]
pmq2['pm penalty']=pmq2['Penalty amount']
pmq2['Month']=6
pmq2.drop('Penalty amount',axis=1,inplace=True)
pmq2

```

```python
df5 = pd.read_excel(lvb, 'PM penalty Q3-21')
df_pmq3=df5.rename(columns=df5.iloc[0]).drop(df5.index[0])
pmq3=df_pmq3[['ATM ID','Penalty Amount']]
pmq3['Month']=9
pmq3['pm penalty']=pmq3['Penalty Amount']
pmq3.drop('Penalty Amount',axis=1,inplace=True)
pmq3
```

```python
pm=pd.merge(pmq2,pmq3,on=['ATM ID','Month'],how='outer')
pm['pm penalty']=pm['pm penalty_x'].fillna(pm['pm penalty_y'])
pm.drop(['pm penalty_x','pm penalty_y'],axis=1,inplace=True)
pm
```

```python
df_equi.info()
df_equi['ATM ID']=df_equi['ATM ID'].astype('object')
```

```python
df_mcu['ATM ID']=df_mcu['SiteName']
df_mcu.drop('SiteName',axis=1,inplace=True)
```

```python
mcu_eq=pd.merge(df_mcu,df_equi,on=['ATM ID','Month'],how="outer")
mcu_eq
```

```python
mcu_eq_footage=pd.merge(mcu_eq,footage,on=['ATM ID','Month'],how="outer")
mcu_eq_footage
```

```python
mcu_eq_footage_pm=pd.merge(mcu_eq_footage,pm,on=['ATM ID','Month'],how="outer")
mcu_eq_footage_pm
```

```python
mcu_eq_footage_pm=mcu_eq_footage_pm.fillna(0)
```

```python
df_all=mcu_eq_footage_pm.groupby(['ATM ID','Month']).sum().reset_index()
df_all
```

```python
dropq1=df_all[(df_all["Month"]==8) | (df_all["Month"]==9)]
df_all.drop(dropq1.index,inplace=True)
df_all
```

```python
df_all['Month'].replace([4,5,6,7],["Apr","May","June","July"],inplace=True)
df_all.to_csv('LVBPenaltyAprToJuly2.csv')
```
